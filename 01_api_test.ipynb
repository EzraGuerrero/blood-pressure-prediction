```python
# Test a fake API

import requests

url = "https://www.w3schools.com/python/demopage.js"
response = requests.get(url)
print(response.status_code) # Returns a number that indicates the status (200 is OK, 404 is Not Found)

# Check if successful
if response.status_code == 200:
    print("Success!")
    data = response.json() # .json() method returns a JSON object of the result (if the result was written in JSON format, if not it raises an error)
    print(data)
else:
    print(f"Error: {response.status_code}")
```
