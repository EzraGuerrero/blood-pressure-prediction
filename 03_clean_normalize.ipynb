# Load Data

```python
import pandas as pd
import os

# Load .xpt file
demo = pd.read_sas("data/raw/DEMO_L.xpt")
bpx = pd.read_sas("data/raw/BPXO_L.xpt")

print(demo.head())
print(bpx.head())

# Save as CSV for future use
os.makedirs("data/processed", exist_ok=True)
demo.to_csv("data/processed/demo_clean.csv" , index=False)
bpx.to_csv("data/processed/bpx_clean.csv" , index=False)
```

# Cleaning Steps:
- Merged demo + BPX on `SEQN`
- Dropped missing values
- Scaled age and BP using StandardScaler
- Saved to `cleaned_normalized.csv`

```python
# Clean & Normalize NHANES Data

import pandas as pd

# 1) Load data:

demo = pd.read_csv("data/processed/demo_clean.csv")
bpx = pd.read_csv("data/processed/bpx_clean.csv")

# 2) Clean: Handle Missing Values:

# drop rows with missing BP or age
cleaned = demo.merge(bpx, on="SEQN", how="inner")
cleaned = cleaned.dropna(subset=["BPXOSY1", "BPXODI1", "RIAGENDR", "RIDAGEYR"])

# keep only relevant columns
cleaned = cleaned[["SEQN", "RIDAGEYR", "RIAGENDR", "BPXOSY1", "BPXODI1"]]

# 3) Normalize: Scale Age & BP

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cleaned[["RIDAGEYR_scaled","BPXOSY1_scaled", "BPXODI1_scaled"]] = scaler.fit_transform(cleaned[["RIDAGEYR", "BPXOSY1", "BPXODI1"]])

# 4) Save Cleaned Data:

cleaned.to_csv("data/processed/cleaned_normalized.csv", index=False)
```
